```python
Numbers=np.loadtxt("ADD.txt",skiprows=1)
```

```python
A=Numbers[:,0]
B=Numbers[:,1]
C=Numbers[:,2]
Cout=Numbers[:,3]
SumA=Numbers[:,4]
```

```python
mm=[]
counter=0
for i,j,k,m,z in zip(A,B,C,SumA,range(len(A))):
    if i+j+k!=m:
        mm.append((z,m))
        counter +=1
        print(f'There is an error {i}+{j}+{k} not equal to {m} at index # {z}')
if len(mm)==0:
    print(f'No Error in this design')
else:
    print(f'There are {counter} errors in this design')    
```
