```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-2*np.pi, 2 * np.pi, 100)
```

```python
Ya = np.sin(X)
Yb = np.cos(X)
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
def plot_slope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys / Xs)
X = np.linspace(-3, 3, 100)
Y = np.exp(-X ** 2)
plt.plot(X, Y,'r')
plot_slope(X, Y)
plt.show()
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.sin(X)
Ya[Ya[:]>0.5]=0.5
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
plt.plot(X, Yb,'k')
plt.plot(X, Ya,'k--')
plt.fill_between(X,Ya , Yb, color='red', alpha='0.1')
#plt.grid()
#plt.grid(b=True, which='major', color='#666666', linestyle='-.')
plt.grid(b=True, which='major', color='b', linestyle='-', alpha=0.2)
plt.savefig('questions.png')
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.sin(X)
Yc = np.sin(X)
Ya[Ya[:]>0]=0
Yc[Yc[:]<0]=0
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.plot(X, Yc)
plt.plot(X, Yb,'k')
plt.plot(X, Ya,'k--')
plt.plot(X, Yc,'k--')
plt.plot(X, Yb,'k')
plt.fill_between(X,Ya , Yb, color='red', alpha='0.5')
plt.fill_between(X,Yc , Yb, color='blue', alpha='0.5')

#plt.grid()
#plt.grid(b=True, which='major', color='#666666', linestyle='-.')
plt.grid(b=True, which='major', color='#999999', linestyle='-', alpha=0.2)
plt.savefig('questions1.png')
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Ya = 2*np.sin(X)
Yb = np.sin(X)
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.plot(X, Yb,'k')
plt.plot(X, Ya,'k--')
plt.plot(X, Yb,'k')
plt.fill_between(X,Ya , Yb, color='red', alpha='0.25')
plt.grid(b=True, which='major', color='#999999', linestyle='-', alpha=0.2)
plt.savefig('questions2.png')
```
