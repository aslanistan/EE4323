# Plotting in Python

This lecture we will use basics of plotting in Python using Matplotlib. 

We will need to load matplotlib pyplot module first. You can do it couple different ways
```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```
Assume we collected Electrical Engineering department student numbers for years 2012 through 2018. Assume 
Years=[2012,2013,2014,2015,2016,2017,2018]
Students=[80,88,72,85,102,109,122]

````python
import matplotlib.pyplot as plt
Years=[2012,2013,2014,2015,2016,2017,2018]
Students=[80,88,72,85,102,109,122]
plt.plot(Years, Students)
plt.show()
````

### plt.plot

Before we start going more detail in different types of plot, we may need to look at how to modify the plot that is shown above.

````python
import matplotlib.pyplot as plt
#plt.rcdefaults()
plt.style.use('default')
plt.rc('lines', linewidth=2,marker='o',color='r')
Labels=[2012,2013,2014,2015,2016,2017,2018]
Years=[2012,2013,2014,2015,2016,2017,2018]
Students=[80,88,72,85,102,109,122]
plt.plot(Years, Students)
#plt.xticks(Years, Labels, rotation='vertical')
plt.xticks(Years, Labels, rotation=180)
plt.margins(0.1)
plt.grid()
plt.show()
````

## Title, x and y axis labels

Noe we can put title and x,y axis labels. 

````python
plt.xlabel(r'$\int\ Y^2\ dt\ \ [V^2 s]$',fontweight='semibold')
plt.ylabel('Number of Students',fontsize=18,labelpad=40)
plt.title('Electrical Engineering Students\n EE only')
````

````python
locs = ['center', 'left', 'right']
plt.xlabel('Years',fontweight='semibold')
plt.ylabel('Number of Students',fontsize=18,labelpad=40)
plt.title('Electrical Engineering Students\n EE only',pad=40,loc=locs[0])
````

````python
plt.text(.1,.4,"Some text",fontstyle="italic",fontweight="bold",rotation=90,color="red",family="serif",bbox=dict(boxstyle = "round",
                  facecolor = "purple",alpha=0.2))

plt.xlabel('Years',rotation=45)
plt.ylabel('Number of Students')
plt.title('Electrical Engineering Students\n EE only')
plt.margins(0.1)
````

````python
plt.text(.1,.4,"Some text",fontstyle="italic",fontweight="bold",rotation=90,color="red",family="serif",bbox=dict(boxstyle = "round",
                  facecolor = "purple",alpha=0.2))
plt.annotate(
    "Here is the text",
    (.15, .5),
    (.5, .3),
    arrowprops=dict(arrowstyle = "->"))
plt.xlabel('Years',rotation=45)
plt.ylabel('Number of Students')
plt.title('Electrical Engineering Students\n EE only')
plt.margins(0.1)
````

````python
plt.rcParams['text.usetex'] = True
plt.text(
    0.5, 0.5,
    r"$f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}}" +
    r"e^{- \frac{(x - \mu)^2}{2\sigma^2}}$")
plt.xlabel('Years',rotation=45)
plt.ylabel('Number of Students')
plt.title('Electrical Engineering Students\n EE only')
plt.margins(0.1)
````

````python
# Plotting
# Simple Plot

from matplotlib import pyplot as plt

Years = [1991,1992,1993,1994,1995]
CEE = [20, 21, 25, 27, 12]

plt.plot(Years, CEE, label='Computer')
plt.xticks(Years)
#plt.yticks(Students)
plt.yticks(range(12,31))
plt.xlabel('Years')
plt.ylabel('Students')
plt.title('EE Students')

plt.legend()
plt.tight_layout()

plt.savefig('plot.png')

plt.show()
````
